```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/Unified Dataset/unified_dataset_with_hybrid_target.csv")

df.shape
```

```python
missing_rate = df.isna().mean()

high_missing_cols = missing_rate[missing_rate > 0.90].index.tolist()

high_missing_cols
```

```python
for col in high_missing_cols:
    if df[col].dtype == "object":
        print(f"\n===== {col} =====")
        print(pd.crosstab(df[col], df["target_hyper_hybrid"], normalize="columns"))
```
